```python
# Define lists of positive and negative words
positive_words = ["good", "great", "excellent", "wonderful", "amazing"]
negative_words = ["bad", "terrible", "horrible", "awful", "poor"]

# Function to classify the document based on polarity words
def classify_sentiment(document):
    # Tokenize the document into words (you may need a more advanced tokenizer)
    words = document.split()

    # Initialize counts for positive and negative words
    positive_count = 0
    negative_count = 0

    # Count the positive and negative words in the document
    for word in words:
        if word in positive_words:
            positive_count += 1
        elif word in negative_words:
            negative_count += 1

    # Determine the overall sentiment
    if positive_count > negative_count:
        return "positive"
    elif negative_count > positive_count:
        return "negative"
    else:
        return "neutral"

# Example document
document = "The movie was really good and the acting was excellent."

# Classify the sentiment of the document
sentiment = classify_sentiment(document)
print(f"Document sentiment: {sentiment}")
```
